# 1. The Geometry of Linear Equations

## 1. Row Picture (Geometry of Equations)

### Choose an x-range for plotting

```python
import numpy as np
import pylab as pl

x_range = np.linspace(-5, 5, 50)
x_range
```

### Express each equation as y = f(x)
1.  2x + y = 5
2. x - y = 1

```python
y1 = 5 - 2 * x_range
y2 = x_range - 1
```

```python
y1[:3]
```

```python
y2[:3]
```

### Plot both lines in 2D

```python
import matplotlib.pyplot as plt

plt.plot(x_range, y1, label='y1')
plt.plot(x_range, y2, label='y2')
plt.legend()
plt.show()
```

### Now find the intersection algebraically
#### Ax = b

```python
A = np.array([[2, 1], [1, -1]])
b = np.array([5, 1])

print(f"{A}x = {b}")
```

### Solve the system using NumPy

```python
solution = np.linalg.solve(A, b)
solution
```

### Mark the intersection point on your plot

```python
plt.plot(x_range, y1, label='y1')
plt.plot(x_range, y2, label='y2')
plt.scatter(2, 1, color='red')
plt.text(1.8, 2, "(2, 1)")
plt.legend()
plt.show()
```

## 2. The Column Picture

### Extract the columns of A

```python
col1 = A[:, 0]
col2 = A[:, 1]
print(col1, col2)
```

### Form the linear combination using your solution x = 2, y = 1

```python
combo = 2 * col1 + 1 * col2
combo
```

### Plot the columns and b as 2D vectors

```python
plt.quiver(0, 0, col1[0], col1[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(0, 0, col2[0], col2[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, combo[0], combo[1], angles='xy', scale_units='xy', scale=1, color='green')
plt.text(2, 1, "(2, 1)")
plt.text(0, -1, "(0, -1)")
plt.text(5, 1, "(5, 1)")

plt.xlim(-2, 6)
plt.ylim(-3, 4)
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

### Check the ranks of A and the augmented matrix [A|b]

A=[2 4 - 1 2]

b=[5 12]

[Ab]=[2 4 - 1 2 - 5 12]

```python
A = np.array([[2, 4], [1, 2]])
b = np.array([5, 12])

rank_a = np.linalg.matrix_rank(A)
rang_aug = np.linalg.matrix_rank(np.column_stack([A, b]))
print(rank_a, rang_aug)
```
